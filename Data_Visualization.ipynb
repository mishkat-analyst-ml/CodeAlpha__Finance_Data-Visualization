```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df=pd.read_csv('Finance_data.csv')
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#check strip whitespace in column names
df.columns = df.columns.str.strip()
```

```python
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['age'], bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.countplot(x='Investment_Avenues', data=df)
plt.title('Investment Avenues Distribution')
plt.xlabel('Investment Avenues')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='gender',hue='Investment_Avenues', data=df)
plt.title('Investment Avenues Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(y='Duration', data=df)
plt.title('Duration Distribution')
plt.show()
```

```python
sns.countplot(y='Expect', data=df)
plt.title('Expectation Distribution')
plt.show()
```

```python
sns.countplot(y='Reason_Equity', data=df)
plt.title('Reason for Investing in Equity Distribution')
plt.show()
```

```python
sns.countplot(x='Reason_FD', data=df)
plt.title('Reason for Investing in FD Distribution')
plt.show()
```

```python
sns.countplot(y='Source', data=df)
plt.title('Source of Financial Information Distribution')
plt.show()
```

```python
sns.boxplot(x="Investment_Avenues", y="age", data=df)
plt.title('Age Distribution by Investment Avenues')
plt.xlabel('Investment Avenues')
plt.ylabel('Age')
plt.show()
```
